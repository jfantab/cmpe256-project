## Reading in File

```python
raw_genres = []
with open('/content/drive/MyDrive/cmpe256-project/million_songs_data/genres/msd_tagtraum_cd1.cls', 'r') as fp:
    raw_genres = fp.readlines()
    raw_genres = raw_genres[7:] # Ignoring commented lines
len(raw_genres)
```

## Parsing

```python
genres = [row.strip().split('\t') for row in raw_genres]
```

```python
genres_dict = dict()

for row in genres:
    song_id = row[0]
    genres = row[1:]

    genres_dict[song_id] = genres
```

```python
genres_dict
```

## One-Hot Encoding Genres

```python
genres_list = set()
for tup in genres_dict.values():
    for ele in tup:
        if ele not in genres_list:
            genres_list.add(ele)

genres_list = list(genres_list)

print(len(genres_list))
genres_list
```

```python
import csv

fp = open("/content/drive/MyDrive/cmpe256-project/million_songs_data/genres/song_genre_one_hot_encodings.csv", "w")

columns = ["track_id"]
columns.extend(genres_list)

writer = csv.DictWriter(fp, fieldnames=columns)
writer.writeheader()

for song_id, song_genres in zip(genres_dict.keys(), genres_dict.values()):
    tmp = {}

    tmp["track_id"] = song_id

    for index, genre in enumerate(genres_list):
        if genre in song_genres:
            tmp[genre] = 1
        else:
            tmp[genre] = 0

    writer.writerow(tmp)
    
fp.close()
```
